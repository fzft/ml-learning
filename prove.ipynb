```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def plot_sigmoid(w, b):
    # Generate x values
    x = np.linspace(-10, 10, 400)
    
    # Compute z = wx + b for each x
    z = w * x + b
    
    # Compute sigmoid values for each z
    y = sigmoid(z)

    # Plot the sigmoid function
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f'Sigmoid Function for w={w} and b={b}', color='blue')
    plt.title(f'Sigmoid Function (w={w}, b={b})')
    plt.xlabel('x')
    plt.ylabel('sigmoid(wx+b)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Change these values to see how the plot changes
w = 9
b = 44
plot_sigmoid(w, b)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def feed_forward(x, w, b, wo, bo):
    # Outputs from the hidden neurons
    h = [sigmoid(wi * x + bi) for wi, bi in zip(w, b)]
    
    # Weighted sum of outputs
    y = sum([wo[i] * hi for i, hi in enumerate(h)]) + bo
    
    return y

# Define weights and biases for the hidden neurons
# Using 4 hidden neurons to approximate the step function
w = [100, 45]
b = [-16, -24]

# Weights and bias for output neuron
wo = [.4, 0.7]
bo = 0

# Plotting
x_values = np.linspace(0, 1, 400)
y_values = [feed_forward(x, w, b, wo, bo) for x in x_values]
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label="Output of Neural Network", color='blue')
plt.title('Neural Network Approximating Step Function')
plt.xlabel('x')
plt.ylabel('Output')
plt.grid(True)
plt.show()
```
