```python
import numpy as np
import matplotlib.pyplot as plt

# Time variable (t) from 0 to 2 with 1000 points
t = np.linspace(0, 2, 1000)

# Calculating the function sin(2πt) + sin(4πt) + sin(6πt)
y = np.sin(2*np.pi*t) + np.sin(4*np.pi*t) + np.sin(6*np.pi*t)

# Plotting the function
plt.figure(figsize=(10, 6))
plt.plot(t, y)
plt.title('Plot of sin(2πt) + sin(4πt) + sin(6πt)')
plt.xlabel('t')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
