```python
import numpy as np
import matplotlib.pyplot as plt

# Time variable (t) from 0 to 2 with 1000 points
t = np.linspace(0, 2, 1000)

# Calculating the function sin(2t) + sin(4t) + sin(6t)
y = np.sin(2*np.pi*t) + np.sin(4*np.pi*t) + np.sin(6*np.pi*t)

# Plotting the function
plt.figure(figsize=(10, 6))
plt.plot(t, y)
plt.title('Plot of sin(2t) + sin(4t) + sin(6t)')
plt.xlabel('t')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
A = 2

phi = np.pi / 4  # Example phase shift (45 degrees)

# Different angular frequencies for comparison
omega_values = [1, 3, 5]  # Different values of omega

# Plotting the signal with different frequencies
plt.figure(figsize=(12, 6))

for omega in omega_values:
    x_t = A * np.cos(omega * t + phi)
    plt.plot(t, x_t, label=f' = {omega}')

plt.title('Effect of Changing Angular Frequency on Cosine Wave')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Creating a discrete time vector
n = np.arange(0, 50)  # Discrete time indices from 0 to 49

# Plotting the discrete signal with different frequencies
plt.figure(figsize=(12, 6))

omega = 1  # Example angular frequency
x_n = A * np.cos(omega * n + phi)
plt.stem(n, x_n, label=f' = {omega}', basefmt=" ")

plt.title('Effect of Changing Angular Frequency on Discrete Cosine Wave')
plt.xlabel('Sample Index (n)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Defining a continuous sine wave
f = 1  # Frequency of the sine wave
t_continuous = np.linspace(0, 2, 1000)  # Continuous time vector
x_continuous = np.sin(2 * np.pi * f * t_continuous)

# Sampling the continuous signal
f_s = 5  # Sampling frequency in Hz
sampling_interval = 1 / f_s  # Interval between samples
n_samples = np.arange(0, 2, sampling_interval)  # Sample points
x_samples = np.sin(2 * np.pi * f * n_samples)  # Sampled signal

# Plotting both continuous and sampled signals
plt.figure(figsize=(12, 6))
plt.plot(t_continuous, x_continuous, label='Continuous Signal')
plt.stem(n_samples, x_samples, 'r', markerfmt='ro', basefmt=" ", label='Sampled Signal')
plt.title('Sampling a Continuous Signal')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```
