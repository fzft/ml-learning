```python
import numpy as np
import matplotlib.pyplot as plt

# Define the mean and covariance matrices for the bivariate Gaussian distributions
mean = np.array([0, 0])

# Covariance matrices
covariances = {
    'a': np.array([[1, 0], [0, 1]]),
    'b': np.array([[1, 1/2], [1/2, 1]]),
    'c': np.array([[1, -1], [-1, 3]])
}

# Create a figure with subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot each bivariate Gaussian distribution
for i, (key, cov) in enumerate(covariances.items()):
    # Calculate the bivariate Gaussian distribution
    rv = np.random.multivariate_normal(mean, cov, size=(10000,))
    
    # Plot the contour plot
    axes[i].set_title(f'({key})  = {cov}')
    axes[i].set_xlabel('x1')
    axes[i].set_ylabel('x2')
    axes[i].hexbin(rv[:,0], rv[:,1], gridsize=50, cmap='inferno', bins='log')
    axes[i].axis('equal')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
